```python
import cv2
import pandas as pd
import glob
import os
from ultralytics import YOLO
from tracker import *

model = YOLO('yolov8s.pt')

def imgwrite(img, dest_path):
    filename = str(s) + '.jpg'
    cv2.imwrite(os.path.join(dest_path, filename), img)

my_file = open("coco.txt", "r")
data = my_file.read()
class_list = data.split("\n")

folder_path = '/content/drive/MyDrive/EssentialTremorProject/closed/patient34 '

for file_name in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file_name)
    print('dossier 0', file_path)
    #if '0' in file_name:
    if file_name == '0°':
        mp4_files = glob.glob(file_path + '/*.mp4')

        dest_path_base = os.path.join(folder_path, '0-crops')  # Base path for all crops folders
        os.makedirs(dest_path_base, exist_ok=True)  # Create base crops folder

        for file_path_mp4 in mp4_files:
            #video_name = os.path.splitext(os.path.basename(file_path_mp4))[0]  # Extract video name
            file_name = file_path_mp4.split('/')[-1]
            print('9ra esm l video')
            if file_name.split('-')[-1] == 'L.mp4':
                area = [(748, 1076), (785, 600), (873, 604), (959, 1077)]
                print('area mrigla L')
            elif file_name.split('-')[-1] == 'R.mp4':
                area = [(1020, 1077), (1075, 751), (1175, 713), (1434, 1077)]
                print('area mrigla R')

            dest_path0 = file_name.split('.')[0] + '-crops'
            dest_path = os.path.join(dest_path_base, dest_path0)  # Create destination path with subfolder
            os.makedirs(dest_path, exist_ok=True)  # Create the subfolder
            print(dest_path + " New empty folder created.")

            print(file_path_mp4 + ' source')
            cap = cv2.VideoCapture(file_path_mp4)

            count = 0
            tracker = Tracker()
            s = 0

            while True:
                ret, frame = cap.read()
                if not ret:
                    break
                count += 1
                if count % 2 != 0:
                    continue

                results = model.predict(frame)

                a = results[0].boxes.data
                px = pd.DataFrame(a).astype("float")

                list = []
                for index, row in px.iterrows():
                    x1 = int(row[0])
                    y1 = int(row[1])
                    x2 = int(row[2])
                    y2 = int(row[3])
                    d = int(row[5])
                    c = class_list[d]
                    if 'person' in c:
                        list.append([x1, y1, x2, y2])

                bbox_idx = tracker.update(list)
                for bbox in bbox_idx:
                    x3, y3, x4, y4, id = bbox
                    x3 -= 10
                    y3 -= 10
                    x4 += 10
                    y4 += 10

                    crop = frame[y3:y4, x3:x4]
                    s += 1
                    if not crop.size == 0:
                        imgwrite(crop, dest_path)

                if cv2.waitKey(1) & 0xFF == 27:
                    break

            cap.release()
            cv2.destroyAllWindows()
```
